```python
# build up library

up = pd.read_csv('msigdb_gene_sets/msigdb_up_mapped_to_HG_U133A.gmt', header=None)

i = 1

lib_up = pd.DataFrame()

for index in up.index:
    lst = []
    print('\r', str(i)+' Out of '+ str(len(up.index)), end=' ', flush=True)
    
    lst.append('_'.join(up.ix[index, 0].split('\t')[0].split('_'))[0:-3])
    
    genes = up.ix[index, 0].split('\t')[2:]
    
    for gene in genes:
        if gene in probes_to_genes.index and probes_to_genes.ix[gene, 'SYMBOL'] != 'NaN':
            lst.append(probes_to_genes.ix[gene, 'SYMBOL'])

    temp = pd.DataFrame(data=lst)
    lib_up = pd.concat([lib_up, temp.T])
    i += 1

lib_up.replace(np.nan, ' ', inplace=True)

lib_up.to_csv('cmap_up.gmt', sep='\t', header=None, index=False)
```

```python
# build up library

down = pd.read_csv('msigdb_gene_sets/msigdb_dn_mapped_to_HG_U133A.gmt', header=None)

i = 1

lib_down = pd.DataFrame()

for index in down.index:
    lst = []
    print('\r', str(i)+' Out of '+ str(len(down.index)), end=' ', flush=True)
    
    lst.append('_'.join(down.ix[index, 0].split('\t')[0].split('_'))[0:-3])
    
    genes = down.ix[index, 0].split('\t')[2:]
    
    for gene in genes:
        if gene in probes_to_genes.index and probes_to_genes.ix[gene, 'SYMBOL'] != 'NaN':
            lst.append(probes_to_genes.ix[gene, 'SYMBOL'])

    temp = pd.DataFrame(data=lst)
    lib_down = pd.concat([lib_down, temp.T])
    i += 1

lib_down.replace(np.nan, ' ', inplace=True)

lib_down.to_csv('cmap_down.gmt', sep='\t', header=None, index=False)
```
